```python
#Verificamos si Opsys es solo Windows 10 o si efectivamente hay mÃ¡s sistemas operativos distintos, tras comprobarlo vemos que sÃ­
valores_unicos = train_raw['OpSys'].unique()
valores_unicos
```

```python
train_raw['OpSys'] = train_raw['OpSys'].replace('No', 'NA')
train_raw['OpSys'] = train_raw['OpSys'].replace('Mac OS ', 'macOS ')
train_raw['OpSys'] = train_raw['OpSys'].replace('Chrome OS', 'Chrome')
train_raw['OpSys'] = train_raw['OpSys'].replace('10 S', '10S')
```

```python
train_raw['OpSys'] = train_raw['OpSys'].str.strip().replace('Mac OS', 'macOS')
```

```python
train_raw['OpSys'] = train_raw['OpSys'].replace(r'\s*Mac\s*OS\s*', 'macOS', regex=True)
train_raw['OpSys'] = train_raw['OpSys'].replace('macOSX', 'macOS X')
```

```python
train_raw['OpSys'] = train_raw['OpSys'].replace(r'Windows\s*10\s*S', 'Windows 10S', regex=True)
```

```python
# Dividir la columna 'OpSys' utilizando un espacio en blanco como separador
train_raw[['SO', 'modso']] = train_raw['OpSys'].str.split(' ', expand=True)

```
